```python
# Python SQL toolkit and Object Relational Mapper
import os
import pandas as pd
import sqlalchemy
import sqlite3
```

```python
conn = sqlite3.connect('BEER.db')
cur = conn.cursor()
```

```python
import glob
path = r'C:\Users\ryanj\anaconda\envs\Project_Real\data\mod'
files = glob.glob(path + '/*.csv')
```

```python
beer = []
for file in files:
    beer_database_df = pd.read_csv(file, usecols= [0,1,2,3,4,5], index_col=None, header=0, keep_default_na = False, nrows = 11)
    beer.append(beer_database_df)

beer_production = pd.concat(beer, axis=0, ignore_index=True, sort=False)
beer_df = beer_production.dropna()
```

```python
cols = [ 'Number of Breweries (1)',
       'Total Barrels (3)', 'Taxable Removals (4)',
       'Total Shipped (Exported) (6)']
beer_df2 = beer_df[cols].replace({',': ''}, regex=True)

beer_df3 = beer_df2.rename({'Number of Breweries (1)' : 'clean_brewry_count','Total Barrels (3)': 'clean_total', 'Taxable Removals (4)':'clean_domestic','Total Shipped (Exported) (6)': 'clean_exported'}, axis=1)
beer_df4 = beer_df.join(beer_df3)
```

```python
beer_df5 = beer_df4[['Year', 'Barrels (31 gallons) (2)',  'clean_brewry_count',
       'clean_total', 'clean_domestic', 'clean_exported']].copy()
```

```python
beer_df5.rename(columns = {'Year':'year', 'Barrels (31 gallons) (2)' : 'brewry_size','clean_brewry_count':'brewier_count', 'clean_total': 'total_barrels' , 'clean_domestic' :'domestic_consumption', 'clean_exported':'exported_barrels'}, inplace=True )
```

```python
beer_df5[['brewier_count', 'total_barrels',
       'domestic_consumption', 'exported_barrels']] = beer_df5[['brewier_count', 'total_barrels',
       'domestic_consumption', 'exported_barrels']].apply(pd.to_numeric)
```

```python
beer_df5.groupby(['year','brewry_size']).agg({'brewier_count': sum ,
                                              'total_barrels': sum})
```

```python
beer_df5.to_sql('annual_production', con=conn , if_exists='replace')
```
