```python
# Python SQL toolkit and Object Relational Mapper
import os
import pandas as pd
import sqlalchemy
import sqlite3
```

```python
conn = sqlite3.connect('BEER.db')
cur = conn.cursor()
```

```python
import glob
path = r'C:\Users\ryanj\anaconda\envs\Project_Real\data\mod'
files = glob.glob(path + '/*.csv')
files
```

```python
beer = []
for file in files:
    beer_database_df = pd.read_csv(file, usecols= [0,1,2,3,4,5], index_col=None, header=0, keep_default_na = False, nrows = 11)
    beer.append(beer_database_df)

beer_production = pd.concat(beer, axis=0, ignore_index=True, sort=False)
beer_df = beer_production.dropna()
# beer_df.rename(columns = {'Year':'year', 'Barrels (31 gallons) (2)' : 'brewry_size','Number of Breweries (1)':'brewier_count', 'Total Barrels (3)': 'total_barrels' , 'Taxable Removals (4)' :'domestic_consumption', 'Total Shipped (Exported) (6)':'exported_barrels'}, inplace=True )
beer_df
```

```python
beer_df.columns.values
```

```python
cols = [ 'Number of Breweries (1)',
       'Total Barrels (3)', 'Taxable Removals (4)',
       'Total Shipped (Exported) (6)']
beer_df2 = beer_df[cols].replace({',': ''}, regex=True)

beer_df3 = beer_df2.rename({'Number of Breweries (1)' : 'clean_brewry_count','Total Barrels (3)': 'clean_total', 'Taxable Removals (4)':'clean_domestic','Total Shipped (Exported) (6)': 'clean_exported'}, axis=1)
beer_df4 = beer_df.join(beer_df3)
```

```python
beer_df5 = beer_df4[['Year', 'Barrels (31 gallons) (2)',  'clean_brewry_count',
       'clean_total', 'clean_domestic', 'clean_exported']].copy()
```

```python
yr2007 = beer_df5['year'] == 2007
yr_2007List = beer_df5[yr2007]
yr_2007Prod = yr_2007List.groupby('brewry_size')[['total_barrels']].sum()
yr_2007Prod.columns = ['2007']
yr2008 = beer_df5['year'] == 2008
yr_2008List = beer_df5[yr2008]
yr_2008Prod = yr_2008List.groupby('brewry_size')[['total_barrels']].sum()
yr_2008Prod.columns = ['2008']
yr2009 = beer_df5['year'] == 2009
yr_2009List = beer_df5[yr2009]
yr_2009Prod = yr_2009List.groupby('brewry_size')[['total_barrels']].sum()
yr_2009Prod.columns = ['2009']
yr2010 = beer_df5['year'] == 2010
yr_2010List = beer_df5[yr2010]
yr_2010Prod = yr_2010List.groupby('brewry_size')[['total_barrels']].sum()
yr_2010Prod.columns = ['2010']
yr2011 = beer_df5['year'] == 2011
yr_2011List = beer_df5[yr2011]
yr_2011Prod = yr_2011List.groupby('brewry_size')[['total_barrels']].sum()
yr_2011Prod.columns = ['2011']
yr2012 = beer_df5['year'] == 2012
yr_2012List = beer_df5[yr2012]
yr_2012Prod = yr_2012List.groupby('brewry_size')[['total_barrels']].sum()
yr_2012Prod.columns = ['2012']
yr2013 = beer_df5['year'] == 2013
yr_2013List = beer_df5[yr2013]
yr_2013Prod = yr_2013List.groupby('brewry_size')[['total_barrels']].sum()
yr_2013Prod.columns = ['2013']
yr2014 = beer_df5['year'] == 2014
yr_2014List = beer_df5[yr2014]
yr_2014Prod = yr_2014List.groupby('brewry_size')[['total_barrels']].sum()
yr_2014Prod.columns = ['2014']
yr2015 = beer_df5['year'] == 2015
yr_2015List = beer_df5[yr2015]
yr_2015Prod = yr_2015List.groupby('brewry_size')[['total_barrels']].sum()
yr_2015Prod.columns = ['2015']
yr2016 = beer_df5['year'] == 2016
yr_2016List = beer_df5[yr2016]
yr_2016Prod = yr_2016List.groupby('brewry_size')[['total_barrels']].sum()
yr_2016Prod.columns = ['2016']
yr2017 = beer_df5['year'] == 2017
yr_2017List = beer_df5[yr2017]
yr_2017Prod = yr_2017List.groupby('brewry_size')[['total_barrels']].sum()
yr_2017Prod.columns = ['2017']
```

```python
yrProdConcat78910 = pd.merge(pd.merge(yr_2007Prod, yr_2008Prod, on = 'brewry_size'),pd.merge(yr_2009Prod, yr_2010Prod, on='brewry_size'),on='brewry_size')
yrProdConcat11121314 = pd.merge(pd.merge(yr_2011Prod, yr_2012Prod, on = 'brewry_size'),pd.merge(yr_2013Prod, yr_2014Prod, on='brewry_size'),on='brewry_size')
yrProdConcat151617 = pd.merge(pd.merge(yr_2015Prod, yr_2016Prod, on = 'brewry_size'),yr_2017Prod, on='brewry_size')
```

```python
yrProdConcat = pd.merge(pd.merge(yrProdConcat78910, yrProdConcat11121314, on = 'brewry_size'),yrProdConcat151617, on='brewry_size')
```
